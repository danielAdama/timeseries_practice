```python
# second, minute, hour, day, week, month, year
data_new = pd.DataFrame({
    'time(seconds)': data['Date_Time'].dt.second,
    'time(minutes)': data['Date_Time'].dt.minute,
    'time(hours)': data['Date_Time'].dt.hour,
    'day_of_week': data['Date_Time'].dt.dayofweek,
    'day_name': data['Date_Time'].apply(lambda x: x.day_name()),
    'month': data['Date_Time'].dt.month,
    'month_name': data['Date_Time'].apply(lambda x: x.month_name()),
    'year': data['Date_Time'].dt.year,
    'day_of_year': data['Date_Time'].dt.dayofyear,
    'today': pd.to_datetime(datetime.date.today()),
    'quarter': data['Date_Time'].dt.quarter,
})
data_new.head()
```

```python
data = pd.concat([data, data_new], axis=1)
data.head()
```

```python
data['is_weekday'] = 0

for i in range(0, len(data)):
    if ((data['day_of_week'].loc[i] == 5) | (data['day_of_week'].loc[i] == 6)):
        data['is_weekday'].loc[i] = 0
    else:
        data['is_weekday'].loc[i] = 1
```

## Let's try to get the age of each applicate

```python
agent_data[['Application_Receipt_Date', 'Applicant_BirthDate']].head()
```

```python
agent_data['Application_Receipt_Date'] = pd.to_datetime(agent_data['Application_Receipt_Date'], format='%m/%d/%Y')
agent_data['Applicant_BirthDate'] = pd.to_datetime(agent_data['Applicant_BirthDate'], format='%m/%d/%Y')
agent_data[['Application_Receipt_Date', 'Applicant_BirthDate']].head()
```

```python
(agent_data['Application_Receipt_Date'][0] - agent_data['Applicant_BirthDate'][0]).days
```

```python
age_days = agent_data.apply(lambda x: (x['Application_Receipt_Date'] - x['Applicant_BirthDate']).days, axis=1)
age_days.head()
```

```python
agent_data['age'] = age_days/365
```
